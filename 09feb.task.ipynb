```python
Q1, Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle.
ans= 
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle

# Example 
car = Vehicle("Car", 200, 30)
bike = Vehicle("Bike", 150, 40)

print("Car details:")
print("Name:", car.name_of_vehicle)
print("Max Speed:", car.max_speed)
print("Average Speed:", car.average_of_vehicle)

print("\nBike details:")
print("Name:", bike.name_of_vehicle)
print("Max Speed:", bike.max_speed)
print("Average Speed:", bike.average_of_vehicle)

output=
Car details:
Name: Car
Max Speed: 200
Average Speed: 30

Bike details:
Name: Bike
Max Speed: 150
Average Speed: 40
```

```python
Q2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity.
ANS=
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle

class Car(Vehicle):
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        super().__init__(name_of_vehicle, max_speed, average_of_vehicle)

    def seating_capacity(self, capacity):
        return f"{self.name_of_vehicle} has a seating capacity of {capacity} people."

# Example usage:
car = Car("Car", 200, 30)
print("Car details:")
print("Name:", car.name_of_vehicle)
print("Max Speed:", car.max_speed)
print("Average Speed:", car.average_of_vehicle)

print("\n" + car.seating_capacity(5))

output= 
Car details:
Name: Car
Max Speed: 200
Average Speed: 30

Car has a seating capacity of 5 people.
```

```python
Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.
ans=
Multiple inheritance is a feature in object-oriented programming where a class can inherit attributes and methods from more than one parent class. In other words, a child class can have multiple parent classes, and it inherits the properties and behaviors of all those parent classes.
In Python, multiple inheritance is supported, and it allows you to create complex class hierarchies and reuse code from different sources. However, when using multiple inheritance, you need to be cautious about potential conflicts or ambiguity that may arise due to the inheritance of similar attributes or methods from different parent classes.

code=

class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        pass

class Flyable:
    def fly(self):
        return "I can fly!"

class Swimmable:
    def swim(self):
        return "I can swim!"

# Bird class inherits from Animal and Flyable
class Bird(Animal, Flyable):
    def make_sound(self):
        return "Chirp Chirp!"

# Fish class inherits from Animal and Swimmable
class Fish(Animal, Swimmable):
    def make_sound(self):
        return "Blub Blub!"

# Penguin class inherits from Bird and Swimmable
class Penguin(Bird, Swimmable):
    def make_sound(self):
        return "Quack Quack!"

# Usage:
bird = Bird("Sparrow")
print(f"{bird.name}: {bird.make_sound()}")
print(bird.fly())

fish = Fish("Salmon")
print(f"{fish.name}: {fish.make_sound()}")
print(fish.swim())

penguin = Penguin("Emperor")
print(f"{penguin.name}: {penguin.make_sound()}")
print(penguin.fly())
print(penguin.swim())


output=

Sparrow: Chirp Chirp!
I can fly!
Salmon: Blub Blub!
I can swim!
Emperor: Quack Quack!
I can fly!
I can swim!

```

```python
Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.
Ans=

In Python, getters and setters are methods that provide a way to access and modify the attributes (variables) of a class, while also allowing for additional logic to be executed during these operations. They are used to ensure controlled access to class attributes and maintain encapsulation, which means that the internal representation of the class is hidden from the outside world.

Getter (Accessor):
A getter method is used to retrieve the value of an attribute. It typically has the format get_<attribute_name>(), and it returns the value of the attribute.

Setter (Mutator):
A setter method is used to modify the value of an attribute. It usually has the format set_<attribute_name>(), and it takes a parameter to set the new value for the attribute.

code= 
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age  # Private attribute with double underscores

    # Getter method for 'age'
    def get_age(self):
        return self.__age

    # Setter method for 'age'
    def set_age(self, new_age):
        if new_age >= 0:
            self.__age = new_age
        else:
            print("Age cannot be negative. Please provide a valid age.")

# Create an instance of the Person class
person1 = Person("Alice", 30)

# Get the age using the getter method
print(person1.get_age())  # Output: 30

# Set a new age using the setter method
person1.set_age(35)
print(person1.get_age())  # Output: 35

# Try to set a negative age (this will be rejected by the setter method)
person1.set_age(-5)  # Output: Age cannot be negative. Please provide a valid age.
print(person1.get_age())  # Output: 35 (The age remains unchanged)
```

```python
Q5.What is method overriding in python? Write a python code to demonstrate method overriding.
ANS= 
Method overriding is a feature in object-oriented programming that allows a subclass to provide a specific implementation for a method that is already defined in its superclass. When a method is overridden in the subclass, the subclass version of the method takes precedence over the superclass version.

code=
class Animal:
    def make_sound(self):
        return "Generic animal sound"

class Dog(Animal):
    def make_sound(self):
        return "Bark bark!"

class Cat(Animal):
    def make_sound(self):
        return "Meow meow!"

class Bird(Animal):
    pass  # Bird inherits the make_sound method from Animal

# Create instances of each class
animal = Animal()
dog = Dog()
cat = Cat()
bird = Bird()

# Call the make_sound method for each instance
print("Animal sound:", animal.make_sound())  # Output: Generic animal sound
print("Dog sound:", dog.make_sound())        # Output: Bark bark!
print("Cat sound:", cat.make_sound())        # Output: Meow meow!
print("Bird sound:", bird.make_sound())      # Output: Generic animal sound (inherited from Animal)
```
